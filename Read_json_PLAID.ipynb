```python
#Code to sort appliances from json metadata file to Folder 
#Name of appliances and its mode in if condition and new_path

import json
import os
f = open('meta_2017.json')
data = json.load(f)
for i in data:
    if(i['meta']['appliance']['type'] == "Compact Fluorescent Lamp"): 
       file_name = str(i['id']) + '.csv' 
       old_path = "D:\\LAB\\NILM\\DATA\\2017\\2017\\" + file_name
       new_path = "D:\\LAB\\NILM\\DATA\\SORTED_DATA\\Compact Lamp\\" + file_name
       os.rename(old_path, new_path)
    
f.close()
```
